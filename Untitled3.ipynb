```python
#To combine all the data

mypath = '.\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data'
all_data = pd.DataFrame()
for f in os.listdir(mypath):
    df = pd.read_csv(join(mypath, f))
    all_data = pd.concat([all_data, df])
all_data.to_csv('Python course.csv')   
```

```python
# To delete the na data 
all_data.isna().sum()
#for col in df.columns:
    #data_missing = np.mean(df['col'].isna())
```

```python
#drop null, na value from data tabel
all_data = all_data.dropna(how='all')
```

```python
#To add a month column
all_data["Months"] = all_data['Order Date'].str[0:2]
# To change data format to int32
all_data["Months"] = all_data['Months'].astype("int32")
```

```python
# There is or value in the data. So we want to get rid of that

all_data = all_data[all_data['Months'] != "Or"]
all_data.head()
```

```python
# To calculate the sale with multiply formula
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"])
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"])
all_data["Sales"] = all_data["Quantity Ordered"]*all_data["Price Each"]
all_data.head()
```

```python
#Change the format to longdate form
A =np.arange(1,13)
def convert_to_monthname(Li):
    Months = []
    for i in Li:
        Months.append(calendar.month_name[i])
    return Months
months = convert_to_monthname(A)
```

```python
Sales_per_month = all_data.groupby('Months').sum()['Sales']
```

```python
#To visualize the correlation betweeen month and sales we're gonna use bar chart

plt.bar(months, Sales_per_month)
plt.xlabel("Months")
plt.ylabel('#Sales in USD')
plt.xticks(months, rotation='45', color="r")
plt.show()
```

```python
#Which city has the most sale?

all_data["City"] = all_data["Purchase Address"]

all_data.drop(columns = 'City', inplace=True)
all_data.head()
```

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split( )[0]

all_data["City"] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
```

```python
sales_per_city = all_data.groupby('City').sum()["Sales"]
sales_per_city
```

```python
#To visualize the correlation betweeen month and sales we're gonna use bar chart
cities = [city for city, df in all_data.groupby('City')]
plt.bar(cities,sales_per_city)
plt.xlabel("City")
plt.ylabel('#Sales in USD')
plt.xticks(cities, rotation='vertical', color="r")
plt.show()
```

```python
all_data["Order Date"] = pd.to_datetime(all_data["Order Date"])
all_data["Order Date"].head()
```

```python
all_data["Hours"] = all_data["Order Date"].dt.hour
all_data['Minutes'] = all_data["Order Date"].dt.minute

all_data["Minutes"].head(10)
```

```python
hours = [hour for hour, df in all_data.groupby('Hours')]
plt.plot(hours, all_data.groupby("Hours").count())
plt.xticks(hours)
plt.grid()
plt.show()
```

```python
df = all_data[all_data["Order ID"].duplicated(keep=False)]
df["Grouped"] = df.groupby("Order ID")['Product'].transform (lambda x: ','.join(x))
df = df[["Order ID", "Grouped"]].drop_duplicates()
df["Grouped"].head()
```

```python
count = Counter()
for row in df["Grouped"]:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list, 2)))
count.most_common()   
```

```python
products = [product for product, df in all_data.groupby('Product')]
quantity_ordered = all_data.groupby('Product', as_index=True).sum()['Quantity Ordered']
prices = all_data.groupby("Product", as_index=True).mean()["Price Each"]
```

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, color='r')
ax1.set_xlabel("Prodcut")
ax2.set_ylabel("Price Each")
ax1.set_xticklabels(products, rotation='vertical', size='8')
plt.show()
```
